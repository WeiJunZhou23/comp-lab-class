```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import numpy as np
```

```python
traj = md.load_trr('adp_exchange4temps.trr', top = 'adp_exchange4temps.gro')
```

```python
phi_angles = md.compute_phi(traj)
psi_angles = md.compute_psi(traj)
```

```python
H, xedges, yedges = np.histogram2d(phi_angles[1].flatten(), psi_angles[1].flatten(), bins=50)
H = - np.log(H.T)
H = H - np.min(H)
plt.figure(figsize=(6, 6))
ax=plt.axes()
ax.set_facecolor('blue')
plt.imshow(H, interpolation='nearest', origin='lower', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])
plt.xlabel('Phi')
plt.ylabel('Psi')
# plt.ylim(-5, 5)
# plt.xlim(-5, 5)
plt.title("480K", fontsize=15)
plt.colorbar()
#plt.savefig("Time_vs_E2E_dist_and_rad_gy.png", dpi=500, bbox_inches='tight')
plt.show()
```
