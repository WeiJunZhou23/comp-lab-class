```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df_300 = pd.read_csv('energy_300.xvg', sep='\s+', skiprows=24, header=None)
df_360 = pd.read_csv('energy_360.xvg', sep='\s+', skiprows=24, header=None)
df_420 = pd.read_csv('energy_420.xvg', sep='\s+', skiprows=24, header=None)
df_480 = pd.read_csv('energy_480.xvg', sep='\s+', skiprows=24, header=None)
```

```python
weight_300 = np.ones_like(df_300[1]) / len(df_300[1])
weight_360 = np.ones_like(df_360[1]) / len(df_360[1])
weight_420 = np.ones_like(df_420[1]) / len(df_420[1])
weight_480 = np.ones_like(df_480[1]) / len(df_480[1])

plt.figure()

plt.hist(df_300[1], weights=weight_300, edgecolor='black', color='red', label='T=300 K')
plt.hist(df_360[1], weights=weight_360, edgecolor='black', color='yellow', label='T=360 K')
plt.hist(df_420[1], weights=weight_420, edgecolor='black', color='blue', label='T=420 K')
plt.hist(df_480[1], weights=weight_480, edgecolor='black', color='green', label='T=480 K')

plt.legend()
plt.title("Potential Energy Histogram")
plt.xlabel("Potential (kJ/mol)", fontsize=10)
plt.ylabel("Frequency", fontsize=10)
plt.savefig('Potential_Energy.png', dpi=500, bbox_inches='tight')
plt.show()
```
