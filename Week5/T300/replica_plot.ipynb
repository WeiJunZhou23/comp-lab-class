```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('replica_temp.xvg', sep='\s+', header=None)
```

```python
for i in [1, 2, 3, 4]:
    plt.figure()
    plt.plot(df[0][::200], df[i][::200], color='black')
    plt.title("replica %s" % i)
    plt.xlabel("Time, ps", fontsize=10)
    plt.savefig('replica_%s.png' % i, dpi=500, bbox_inches='tight')
    plt.show()
```
