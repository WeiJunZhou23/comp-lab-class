```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('energy.xvg', sep='\s+', skiprows=24, header=None)
df.head(10)
```

```python
plt.figure()
plt.plot(df[0], df[1], color='black')
plt.xlabel("Time (ps) ", fontsize=10)
plt.ylabel("Potential (kJ/mol)", fontsize=10)
plt.xlim(0, 5000)
#plt.savefig('replica_%s.png' dpi=500, bbox_inches='tight')
plt.show()
```
