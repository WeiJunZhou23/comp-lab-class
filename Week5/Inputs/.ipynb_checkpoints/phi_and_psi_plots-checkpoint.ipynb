```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import numpy as np
```

```python
traj = md.load_trr('adp.trr', top = 'adp.gro')
```

```python
phi_angles = md.compute_phi(traj)
psi_angles = md.compute_psi(traj)
```

```python
H, xedges, yedges = np.histogram2d(psi_angles[1].flatten(), phi_angles[1].flatten(), bins=100)
H = - np.log(H.T)
H = H - np.min(H)
plt.figure()
plt.imshow(H, interpolation='nearest', origin='lower', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])
plt.xlabel('phi')
plt.ylabel('psi')
plt.title("histogram", fontsize=15)
plt.colorbar()
#plt.savefig("Time_vs_E2E_dist_and_rad_gy.png", dpi=500, bbox_inches='tight')
plt.show()
```
