```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_1 = pd.read_csv('msd_oxygen.xvg', skiprows=20, sep='\s+', header=None)
df_2 = pd.read_csv('msd_Na.xvg', skiprows=20, sep='\s+', header=None)
df_3 = pd.read_csv('msd_Cl.xvg', skiprows=20, sep='\s+', header=None)
```

```python
plt.figure()
plt.plot(df_1[0], df_1[1], color='black', label='Oxygen')
plt.plot(df_2[0], df_2[1], color='red', label='Na')
plt.plot(df_3[0], df_3[1], color='blue', label='Cl')
plt.xlabel("Time(ps)", fontsize=10)
plt.ylabel("MSD (nm^2)", fontsize=10)
plt.legend()
plt.title("Mean Square Displacement", fontsize=15)
plt.savefig("MSD.png", dpi=500, bbox_inches='tight')
plt.show()
```
