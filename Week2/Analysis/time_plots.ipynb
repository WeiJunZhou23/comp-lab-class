```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../Data/1hz3_T310.run.25000000.energy.xvg', skiprows=29, sep='\s+')
df.columns = ['Time', 'Potential', 'Kinetic Energy', 'Total Energy', 'Temperature', 'Pressure', 'Volume']
df
```

```python
x_val = list(df['Time']/1000)
y_titles = ['Potential', 'Kinetic Energy', 'Total Energy', 'Temperature', 'Pressure', 'Volume']
y_labels = ['Potential, kJ/mol', 'Kinetic Energy, kJ/mol', 'Total Energy, kJ/mol', 'Temperature, K', 
            'Pressure, bar', 'Volume, nm^3']

i=0

for y_title in y_titles:
    y_val = list(df[y_title])
    plt.figure()
    plt.plot(x_val, y_val, color='black')
    plt.xlabel("Time, ns", fontsize=10)
    plt.ylabel("%s" % y_labels[i], fontsize=10)
    plt.xlim(0, 50)
    plt.title("Time vs %s" % y_title, fontsize=15)
    plt.savefig("Time_vs_%s.png" % y_title, dpi=500, bbox_inches='tight')
    plt.show()
    
    i += 1
```

```python
x_val = list(df['Time']/1000)
y_val_potential = list(df['Kinetic Energy'])
y_val_temp = list(df['Temperature'])

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.set_xlabel("Time, ns", fontsize=10)
ax1.set_ylabel("Kinetic Energy, kJ/mol", fontsize=10)
ax2.set_ylabel("Temperature, K", fontsize=10)

ax1.plot(x_val, y_val_potential, color='yellow', label='Kinetic Energy')
ax2.plot(x_val, y_val_temp, color='green', label='Temperature')
plt.xlim(0, 50)
plt.title("Time vs Kinetic Energy & Temperature")
ax1.legend(loc='upper right')
ax2.legend(loc='lower right')
plt.show()
```
