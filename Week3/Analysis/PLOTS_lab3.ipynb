```python
import pandas as pd
from matplotlib import pyplot as plt
import glob, os
```

### Plots for potential, pressure, temperature, and density.

```python
files = glob.glob('*.xvg')
df_col = [['Time', 'Potential'], ['Time', 'Tempareture'], ['Time', 'Pressure'], ['Time', 'Density']
         ]
start_row = [24, 24, 24, 24]
plot_titles = ['Energy_Minimization', 'Temperature',  'Pressure', 'Density']
x_labels = ['Energy Minimization Step', 'Time (ps)',  'Time (ps)', 'Time (ps)']
y_labels = ['Potential Energy (kJ/mol)', 'Temperature (K)',  'Pressure (bar)', 
            'Density (kg/m^3)']
i = 0
for file in files:
    print(file)
    df = pd.read_csv(file, skiprows=start_row[i], sep='\s+', header=None)
    df.columns = df_col[i]
    #print(df.head(5))
    plt.figure()
    plt.plot(df['Time'], df[df_col[i][1]], color='black')
    plt.xlabel(x_labels[i], fontsize=12)
    plt.ylabel(y_labels[i], fontsize=12)
    plt.title(plot_titles[i], fontsize=15)
    plt.savefig("%s.png" % plot_titles[i], dpi=500, bbox_inches='tight')
    plt.show()
    i += 1
```

### RMSD plots

```python
df_1 = pd.read_csv('/scratch/work/courses/CHEM-GA-2671-2022fa/wz2449/Week3_result/PLOTS/rmsd/rmsd_1ns/rmsd.xvg', skiprows=18, sep='\s+', header=None)
df_2 = pd.read_csv('/scratch/work/courses/CHEM-GA-2671-2022fa/wz2449/Week3_result/PLOTS/rmsd/rmsd_1ns/rmsd_xtal.xvg', skiprows=18, sep='\s+', header=None)

df_1.columns = ['Time', 'RMSD']
df_2.columns = ['Time', 'RMSD']

plt.figure()
plt.plot(df_1['Time'], df_1['RMSD'], color='black', label='Equilibrated')
plt.plot(df_2['Time'], df_2['RMSD'], color='red', label='Crystal')
plt.xlabel('Time (ns)', fontsize=12)
plt.ylabel('RMSD (nm)', fontsize=12)
plt.title('RMSD', fontsize=15)
plt.legend(loc=4)
plt.savefig('RMSD_1ns.png', dpi=500, bbox_inches='tight')
plt.show()
```

```python
df_3 = pd.read_csv('/scratch/work/courses/CHEM-GA-2671-2022fa/wz2449/Week3_result/PLOTS/rmsd/rmsd_50ns/rmsd_2.xvg', skiprows=18, sep='\s+', header=None)
df_4 = pd.read_csv('/scratch/work/courses/CHEM-GA-2671-2022fa/wz2449/Week3_result/PLOTS/rmsd/rmsd_50ns/rmsd_xtal.xvg', skiprows=18, sep='\s+', header=None)

df_3.columns = ['Time', 'RMSD']
df_4.columns = ['Time', 'RMSD']

plt.figure()
plt.plot(df_3['Time'], df_3['RMSD'], color='black', label='Equilibrated', linewidth=0.5)
plt.plot(df_4['Time'], df_4['RMSD'], color='red', label='Crystal', linewidth=0.5)
plt.xlabel('Time (ns)', fontsize=12)
plt.ylabel('RMSD (nm)', fontsize=12)
plt.title('RMSD', fontsize=15)
plt.legend(loc=4)
plt.savefig('RMSD_50ns.png', dpi=500, bbox_inches='tight')
plt.show()
```
