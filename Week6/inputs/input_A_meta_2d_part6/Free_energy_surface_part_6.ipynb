```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ala2_metad_biasfactor10_pace500_height1.2_sigma0.3_10ns_2d_fes.dat', sep='\s+', skiprows=9, header=None)
df.columns = ['phi', 'psi', 'free ene', 'der_phi', 'der_psi']
df
```

```python
phi=df['phi'].to_numpy()
psi=df['psi'].to_numpy()
ene=df['free ene'].to_numpy()
```

```python
phi_new=phi[:64]
psi_new=psi[::64]
ene_new=ene.reshape(35,64)
```

```python
plt.figure()
plt.imshow(ene_new, origin='lower', interpolation='nearest', extent=[phi_new[0], phi_new[-1], psi_new[0], psi_new[-1]], cmap='viridis')
plt.xlabel('$\phi$')
plt.ylabel('$\psi$')
plt.title('Free Energy Surface-dialaA-2D')
plt.colorbar(label='Free Energy (kJ/mol)')
plt.savefig('FES_part_6.png')
plt.show()
```
